```python
from solutions.PPO.enviroment import PetriNet
from torchrl.envs.transforms import ActionMask,StepCounter,DTypeCastTransform,Compose,TransformedEnv
import torch

base_env = PetriNet(device="cpu")

transform = Compose([ActionMask(),
                         StepCounter(max_steps=30,step_count_key="step"),
                         DTypeCastTransform(dtype_in=torch.int64,dtype_out=torch.float32,
                                            in_keys="observation",out_keys="observation_f"),])
env = TransformedEnv(base_env,transform)
```

```python
env.rollout(max_steps=10)
```
